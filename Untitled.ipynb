```python
import nltk
from nltk.corpus import stopwords
from collections import Counter
import re
from googletrans import Translator
```

```python
# Initialize the Translator
translator = Translator()
```

```python
def translate_tweet(content):
    try:
        translated = translator.translate(content, dest='en')
        return translated.text
    except Exception as e:
        print(f"Error translating tweet: {e}")
        return content  # Return original if translation fails
```

```python
df['tweet_content'] = df['tweet_content'].apply(translate_tweet)
```

```python
df
```

```python
# Download NLTK stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
# Function to clean and tokenize text
def clean_and_tokenize(text):
    # Remove URLs, hashtags, and special characters
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'#\w+', '', text)
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'[^A-Za-z\s]', '', text)
    words = text.lower().split()
    # Remove stop words
    words = [word for word in words if word not in stop_words]
    return words
```

```python
# Tokenize tweet contents
all_words = []
for tweet in df['tweet_content']:
    all_words.extend(clean_and_tokenize(tweet))
```

```python
# Count word frequencies
word_freq = Counter(all_words)
```

```python
# Display most common words
print(word_freq.most_common(10))
```
