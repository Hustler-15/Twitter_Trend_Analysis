### Translating the tweet content from any other language to english

```python
from googletrans import Translator
```

```python
# Initialize the Translator
translator = Translator()
```

```python
def translate_tweet(content):
    try:
        translated = translator.translate(content, dest='en')
        return translated.text
    except Exception as e:
        print(f"Error translating tweet: {e}")
        return content  # Return original if translation fails
```

```python
# Applying the translation on tweet_content
df['tweet_content'] = df['tweet_content'].apply(translate_tweet)
```

### Correct the mispelled word

```python
from textblob import TextBlob
# Use tqdm to display a progress bar while applying the function
tqdm.pandas(desc="Processing Tweets")
```

```python
# Function to correct spelling in a tweet and count errors
def correct_spelling_and_count_errors(tweet):
    # Create a TextBlob object
    blob = TextBlob(tweet)
    
    # Count the number of words before correction
    original_word_count = len(blob.words)
    
    # Correct spelling
    corrected_tweet = str(blob.correct())
    
    # Create a TextBlob object for the corrected tweet to count errors
    corrected_blob = TextBlob(corrected_tweet)
    
    # Count the number of words after correction
    corrected_word_count = len(corrected_blob.words)
    
    # Calculate the number of errors
    error_count = original_word_count - corrected_word_count
    
    return corrected_tweet, error_count
```
